#merge two data frames

```python
movies=movies.merge(credits,on='title')
```

```python
movies['original_language'].value_counts()
```

```python
#geners
#id
#keywords
#title
#overview
#cast
#crew
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(1)
```

```python
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
print(movies['genres'][0])
```

```python
import ast
ast.literal_eval
```

```python
def convert(obj):
    L= []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
print(movies['genres'][0])
```

```python
movies.head()
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L= []
    counter=0
    for i in ast.literal_eval(obj):
        
        if counter !=3:
             
            L.append(i['name'])
            
            counter+=1
            
        
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L= []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L 
```

```python
 movies['crew']= movies['crew'].apply(fetch_director)
 
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

#remove spaces between all keywords

```python
# Function to remove spaces from list of strings
def remove_spaces_from_list(lst):
    return [i.replace(" ", "") for i in lst] if isinstance(lst, list) else []

# Apply to each column
movies['keywords'] = movies['keywords'].apply(remove_spaces_from_list)
movies['cast']     = movies['cast'].apply(remove_spaces_from_list)
movies['crew']     = movies['crew'].apply(remove_spaces_from_list)
movies['genres']   = movies['genres'].apply(remove_spaces_from_list)

```

```python
movies.head(1)
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
newdf=movies[['movie_id','title','tags']]
```

```python
newdf
```

```python
newdf['tags']=newdf['tags'].apply(lambda x:" ".join(x))
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    
    y=[]
    for i in text.split():
        y.append(ps.stem(i))

    return "  ".join(y)
```

```python
newdf['tags']=newdf['tags'].apply(stem)
```

```python
newdf['tags'][:500]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(newdf['tags']).toarray()
```

```python
cv.get_feature_names_out()[:1000]
```

```python
ps.stem("dancing")
```

```python
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index=newdf[newdf['title']==movie].index[0]
    distance= similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(newdf.iloc[i[0]].title)
    return
```

```python
recommend("Iron Man")
```
